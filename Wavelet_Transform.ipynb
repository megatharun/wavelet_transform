## Simple instruction to use WT library program

---
This program can be used to display wavelet transformation for acoustic emission (AE) data. The program is a work in progress and will include other features for analyzing acoustic emission (AE) data.

The WT library program can be imported and the WT_Display() function can be used to display a 2D image of wavelet transformation data.

```python
from WT import *
```

```python
var1 = WT_Display('V14_2700.csv')
```

```python
var1.imageView()
```
